# Introdução
Este projeto de ciência dos dados tem como objetivo correlacionar os dois indicadores; taxa de usuários de internet e taxa de exportações de alta tecnologia, com o Índice de Desenvolvimento Humano. Mais precisamente, ele pretende responder às perguntas: 

Um país com um IDH alto tem uma taxa maior de usuários de internet? O mesmo ocorre em relação à taxa de exportações de alta tecnologia?

Para realisar o objetivo citado, ferramentas de ánalise de dados; gráficos de dispersão, boxplots e histogramas, noções de mediana e média serão usadas.

# Análise dos dados
Primeramente, os dados serão analisados no ano de 2015. Em seguida serão analisados no periodo de 2001 à 2015.
